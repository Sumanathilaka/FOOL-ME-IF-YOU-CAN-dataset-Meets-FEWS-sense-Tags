```python
with open('fewmapping.txt', 'r',encoding='utf-8') as filemapping:
    mapping=[]
    for i in filemapping:
        senseid=i.split(",")[2].strip()
        mapping.append(senseid)

filemapping.close()
print(mapping)
```

```python
import json
import csv
#evaluating the results
list_original=[]
with open('WSD_1.csv', 'r',encoding='latin-1') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        list_original.append(row['senseid'])
        
print(list_original)

file.close()     
```

```python
list_predicted=[]
with open('Result_of_WSD_1.txt', 'r',encoding='utf-8') as file1:
    for i in file1:
        list_predicted.append(i.strip())

print(list_predicted)
file1.close()
```

```python
from sklearn.metrics import f1_score, confusion_matrix

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

f1_macro = f1_score(list_original, list_predicted, average='macro')
f1_weighted = f1_score(list_original, list_predicted, average='weighted')

# Calculate Confusion Matrix
cm = confusion_matrix(list_original, list_predicted)

plt.figure(figsize=(10, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=mapping, yticklabels=mapping)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Print results
print("Macro-Averaged F1 Score:", f1_macro)
print("Weighted-Averaged F1 Score:", f1_weighted)
print("Confusion Matrix:\n", cm)
```

```python
import os
from pathlib import Path

wsd_path = Path("WSD_Train")

for folder in wsd_path.iterdir():
    if folder.is_dir():
        data_file = folder / "train.data.txt"  #for test data rename it to test
        gold_file = folder / "train.gold.txt"  #for test data rename it to test
        
        if data_file.exists() and gold_file.exists():
            #print(f"Folder: {folder.name}")
            
            with open(data_file, 'r') as data, open(gold_file, 'r') as gold:
                data_lines = data.readlines()
                gold_lines = gold.readlines()
                
                for data_line, gold_line in zip(data_lines, gold_lines):
                    print(data_line.strip(),"\t",gold_line.strip(),"\t",folder.name)                
                    
            
            
```

```python
with open ("fewmapping.txt","r",encoding="utf-8") as file:
    sensedict={}
    for i in file:
        senseunit=i.split(",")[0]
        word=i.split(",")[1].split("_")[0]
        senseid=i.split(",")[2].strip("\n")
        name=str(senseunit)+" "+str(word)
        sensedict[name]=senseid

print(sensedict)
```

```python
import csv

with open('WSD_train.csv', 'r',encoding='latin-1') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        name= str(row['id'])+str(row['name'])
        print(sensedict[name])
```
